```python
df = pd.read_csv('parkinsons.data')
df.head(-5)
```

```python
features = df.loc[:, df.columns != 'status'].values[:,1:]
labels = df.loc[:, df.columns == 'status'].values

sum1 = labels[labels==0].shape[0]
sum2 = labels[labels==1].shape[0]

print("zeros: ", sum1, " " , "ones: ", sum2)
```

```python
scaler = MinMaxScaler((-1,1))
y = labels
x = scaler.fit_transform(features)

x_train,x_test,y_train,y_test=train_test_split(x, y, test_size = 0.2, random_state = 7)
```

```python
model = XGBClassifier()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print(accuracy_score(y_test, y_pred)*100)

input_data = (157.30200,74.99700,0.00145,0.00007,0.00370,0.00554,0.01109,0.04374,0.42600,0.02182,0.03130,0.02971,0.06545,0.02211,21.03300,1,0.414783,0.815285,-4.813031,0.266482,2.301442,0.284654)

```
